### Single Layer Neural network 
-  Goal is to find the function which maps input Celsius to output Fahrenheit
-  The Orginal Formula is given by
$ F^∘ = 1.8*C^∘+32 $ 


#### Importing Modules

```python
import numpy as np
import tensorflow as tf
```

#### Training data

```python
celsius=np.array([-40,-10,0,8,15,22,38],dtype=float)
fahrenheit=np.array([-40,14,32,46,59,72,100],dtype=float)
```

#### Creating a Model
- A single layer with single neuron neural network
- Adding the layer into a Sequential model

```python
layer_0=tf.keras.layers.Dense(units=1,input_shape=[1])
model=tf.keras.Sequential([layer_0])
```

#### Compiling and Training the model

```python
model.compile(loss='mean_squared_error',optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history=model.fit(celsius,fahrenheit,epochs=500,verbose=True)
```

#### Plotting loss curves

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('Number of Epochs')
plt.ylabel('Loss')
plt.plot(history.history['loss'])
```

#### Testing

```python
print(model.predict([100]))
```
